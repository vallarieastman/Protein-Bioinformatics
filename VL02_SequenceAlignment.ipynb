```python
from Bio import AlignIO
from collections import Counter
from IPython.display import HTML, display
import html

# --- Load your alignment ---
aln = AlignIO.read("PF02932_muscle.fasta", "fasta")  # change name/format if needed

# --- Helpers ---
KD = dict(A=1.8,R=-4.5,N=-3.5,D=-3.5,C=2.5,Q=-3.5,E=-3.5,G=-0.4,H=-3.2,
          I=4.5,L=3.8,K=-3.9,M=1.9,F=2.8,P=-1.6,S=-0.8,T=-0.7,W=-0.9,Y=-1.3,V=4.2)

def alignment_window(aln, start, end):
    """Return list[(id, subseq)] for 1-based inclusive window [start, end]."""
    L = aln.get_alignment_length()
    start = max(1, start); end = min(L, end)
    s0, e0 = start-1, end
    return [(rec.id, str(rec.seq[s0:e0])) for rec in aln]

def consensus_of(seqs):
    """Ungapped majority consensus over equal-length strings."""
    L = len(seqs[0])
    cons = []
    for i in range(L):
        col = [s[i] for s in seqs if s[i] != '-']
        cons.append(Counter(col).most_common(1)[0][0] if col else '-')
    return "".join(cons)

def pid_to_cons(seq, cons):
    m = sum(1 for a,b in zip(seq,cons) if a!='-' and b!='-')
    return (sum(1 for a,b in zip(seq,cons) if a==b and a!='-' and b!='-')/m) if m else 0.0

def mean_kd(seq):
    vals = [KD.get(a, 0.0) for a in seq if a != '-']
    return sum(vals)/len(vals) if vals else 0.0

def color_seq_html(seq, cons):
    """Return HTML with per-residue spans: matches tinted, mismatches accented, gaps faint."""
    out = []
    for a,b in zip(seq, cons):
        if a == '-':
            out.append('<span class="gap">-</span>')
        elif b == '-' or a != b:
            # mismatch
            out.append(f'<span class="mismatch">{html.escape(a)}</span>')
        else:
            # match to consensus
            out.append(f'<span class="match">{html.escape(a)}</span>')
    return "".join(out)

def render_window_table(aln, start, end, top_n=15, sort_by="pid"):
    rows = alignment_window(aln, start, end)
    ids, seqs = zip(*rows)
    cons = consensus_of(seqs)

    stats = []
    for rec_id, s in rows:
        stats.append({
            "id": rec_id,
            "seq_html": color_seq_html(s, cons),
            "pid": pid_to_cons(s, cons),
            "kd":  mean_kd(s)
        })
    # sort by %id (default) or hydropathy
    stats.sort(key=(lambda r: (-r["pid"], -r["kd"])) if sort_by=="pid" else (lambda r: (-r["kd"], -r["pid"])))
    stats = stats[:top_n]

    # Build HTML -----THIS IS SOOOOOO COOL! I see why everyone loves Jupyter Notebook now! -------
    width = end - start + 1
    ruler = "".join(str((i+1)%10) for i in range(width))

    css = """
    <style>
    .msa-table {border-collapse: collapse; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; font-size: 12.5px;}
    .msa-table th, .msa-table td {border-bottom: 1px solid #e7e7e7; padding: 6px 8px; vertical-align: top;}
    .msa-table th {background: #f6f8fa; text-align: left;}
    .seq {letter-spacing: 0.5px; white-space: nowrap;}
    .cons {background:#fffbe6;}
    .ruler {color:#666;}
    .match {background: #e7f7ee;}      /* soft green for matches */
    .mismatch {background: #ffecec;}   /* soft red for mismatches */
    .gap {color:#bbb;}
    .meta {color:#333; font-variant-numeric: tabular-nums;}
    .tag {display:inline-block; background:#eef; color:#224; padding:2px 6px; border-radius:6px; margin-left:6px; font-size:11px;}
    </style>
    """

    header = f"""
    <div class="meta"><strong>Window {start}-{end}</strong>
      <span class="tag">length {width}</span>
      <span class="tag">top {len(stats)} seqs</span>
    </div>
    <table class="msa-table">
      <thead>
        <tr><th>Sequence ID</th><th>Fragment</th><th>%ID to cons</th><th>Mean KD</th></tr>
      </thead>
      <tbody>
        <tr>
          <td class="ruler">ruler</td>
          <td class="seq ruler">{ruler}</td>
          <td class="ruler"></td>
          <td class="ruler"></td>
        </tr>
        <tr>
          <td class="cons">consensus</td>
          <td class="seq cons">{color_seq_html(cons, cons)}</td>
          <td class="cons"></td>
          <td class="cons"></td>
        </tr>
    """

    rows_html = "\n".join(
        f'<tr><td>{html.escape(r["id"])}</td>'
        f'<td class="seq">{r["seq_html"]}</td>'
        f'<td class="meta">{r["pid"]*100:5.1f}%</td>'
        f'<td class="meta">{r["kd"]:5.2f}</td></tr>'
        for r in stats
    )

    footer = "</tbody></table>"
    return HTML(css + header + rows_html + footer)

# --- Render a few windows ---
display(render_window_table(aln, 1, 15, top_n=12)) # First slice
display(render_window_table(aln, 16, 30, top_n=12))
display(render_window_table(aln, 31, 45, top_n=12))
display(render_window_table(aln, 100, 115, top_n=12))
```

```python
# ===== Tables of most/least conserved regions (gap-aware) =====
import numpy as np
import pandas as pd
from Bio import AlignIO
from collections import Counter

# -- If you already have these from earlier, keep them; otherwise compute here:
aln = AlignIO.read("PF02932_muscle.fasta", "fasta")

KD = dict(A=1.8,R=-4.5,N=-3.5,D=-3.5,C=2.5,Q=-3.5,E=-3.5,G=-0.4,H=-3.2,
          I=4.5,L=3.8,K=-3.9,M=1.9,F=2.8,P=-1.6,S=-0.8,T=-0.7,W=-0.9,Y=-1.3,V=4.2)

def per_column_conservation(aln, min_coverage=0.7):
    nseq = len(aln); scores = []
    for i in range(aln.get_alignment_length()):
        col = aln[:, i]
        residues = [aa for aa in col if aa != "-"]
        if len(residues) < min_coverage * nseq:
            scores.append(0.0); continue
        counts = Counter(residues)
        scores.append(max(counts.values()) / len(residues))
    return np.asarray(scores, float)

def kd_per_column_mean(aln):
    vals = []
    for i in range(aln.get_alignment_length()):
        kds = [KD[a] for a in aln[:, i] if a in KD]
        vals.append(np.mean(kds) if kds else 0.0)
    return np.asarray(vals, float)

def per_column_coverage(aln):
    nseq = len(aln); L = aln.get_alignment_length()
    cov = np.zeros(L, float)
    for i in range(L):
        col = aln[:, i]
        cov[i] = sum(aa != "-" for aa in col) / nseq
    return cov

def moving_avg(x, k=11):
    x = np.asarray(x, float)
    if k <= 1: return x
    y = np.empty_like(x); r = k//2
    for i in range(len(x)):
        lo, hi = max(0, i-r), min(len(x), i+r+1)
        y[i] = x[lo:hi].mean()
    return y

cons = per_column_conservation(aln, min_coverage=0.7)   # 0..1
kd   = kd_per_column_mean(aln)
cov  = per_column_coverage(aln)                         # 0..1
cons_s = moving_avg(cons, 11)                           # optional, not required below

def window_table(cons, kd, cov, k=15, step=15, min_col_cov=0.7, min_frac_good=0.7):
    """
    Build a table of windows (Start/End/Length/Mean_Conservation/Mean_KD/Mean_Coverage),
    keeping only windows where at least `min_frac_good` of columns have coverage >= `min_col_cov`.
    """
    L = len(cons); rows = []
    for start0 in range(0, L - k + 1, step):
        end0 = start0 + k
        idx = slice(start0, end0)
        cov_slice = cov[idx]
        good_cols = cov_slice >= min_col_cov
        frac_good = good_cols.mean()
        if frac_good < min_frac_good:
            continue  # skip gap-dominated windows
        # compute means on good columns only
        c_vals = cons[idx][good_cols]
        kd_vals = kd[idx][good_cols]
        rows.append({
            "Start": start0 + 1,             # 1-based
            "End": end0,
            "Length": k,
            "Mean_Conservation": float(c_vals.mean()) if len(c_vals) else 0.0,
            "Mean_KD":           float(kd_vals.mean()) if len(kd_vals) else 0.0,
            "Mean_Coverage":     float(cov_slice.mean()),
            "Good_Col_Fraction": float(frac_good),
        })
    return pd.DataFrame(rows)

# Build the table (tweak k/step or coverage thresholds as you like)
df_windows = window_table(cons, kd, cov, k=15, step=15, min_col_cov=0.7, min_frac_good=0.7)

# Most/least conserved (by Mean_Conservation)
top = df_windows.sort_values("Mean_Conservation", ascending=False).head(10)
bottom = df_windows.sort_values("Mean_Conservation", ascending=True).head(10)

# Console-friendly prints (Spyder/terminal)
print("\nMost conserved windows:")
print(top.to_string(index=False, formatters={"Mean_Conservation":"{:0.3f}".format,
                                             "Mean_KD":"{:0.2f}".format,
                                             "Mean_Coverage":"{:0.2f}".format,
                                             "Good_Col_Fraction":"{:0.2f}".format}))

print("\nLeast conserved windows:")
print(bottom.to_string(index=False, formatters={"Mean_Conservation":"{:0.3f}".format,
                                                "Mean_KD":"{:0.2f}".format,
                                                "Mean_Coverage":"{:0.2f}".format,
                                                "Good_Col_Fraction":"{:0.2f}".format}))

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from Bio import AlignIO
from collections import Counter

# ---- load alignment ----
aln = AlignIO.read("PF02932_muscle.fasta", "fasta")
print(f"Loaded alignment: {len(aln)} seqs, length {aln.get_alignment_length()}")

# ---- metrics ----
KD = dict(A=1.8,R=-4.5,N=-3.5,D=-3.5,C=2.5,Q=-3.5,E=-3.5,G=-0.4,H=-3.2,
          I=4.5,L=3.8,K=-3.9,M=1.9,F=2.8,P=-1.6,S=-0.8,T=-0.7,W=-0.9,Y=-1.3,V=4.2)

def per_column_conservation(aln, min_coverage=0.7):
    """Return NumPy array of max residue fraction per column; 0 if coverage < threshold."""
    nseq = len(aln)
    scores = []
    for i in range(aln.get_alignment_length()):
        col = aln[:, i]
        residues = [aa for aa in col if aa != "-"]
        if len(residues) < min_coverage * nseq:
            scores.append(0.0)
            continue
        counts = Counter(residues)
        scores.append(max(counts.values()) / len(residues))
    return np.asarray(scores, dtype=float)

def kd_per_column_mean(aln):
    """Return NumPy array of mean KD per column across all non-gap residues."""
    vals = []
    for i in range(aln.get_alignment_length()):
        kds = [KD[a] for a in aln[:, i] if a in KD]
        vals.append(np.mean(kds) if kds else 0.0)
    return np.asarray(vals, dtype=float)

def smooth(x, k=11):
    """Centered moving average (NumPy in/out)."""
    x = np.asarray(x, dtype=float)
    if k <= 1:
        return x
    y = np.empty_like(x)
    r = k // 2
    for i in range(len(x)):
        lo, hi = max(0, i-r), min(len(x), i+r+1)
        y[i] = x[lo:hi].mean()
    return y

# ---- compute signals (NumPy arrays) ----
cons   = per_column_conservation(aln, min_coverage=0.7)
kd     = kd_per_column_mean(aln)
cons_s = smooth(cons, 11)
kd_s   = smooth(kd,   11)

print(f"Conservation smooth: min={cons_s.min():.2f} med={np.median(cons_s):.2f} max={cons_s.max():.2f}")
print(f"KD (mean)  smooth:   min={kd_s.min():.2f} med={np.median(kd_s):.2f} max={kd_s.max():.2f}")

# ---- domain calling (adaptive thresholds) ----
def call_domains_mask(mask, min_len=12):
    spans = []
    i, L = 0, len(mask)
    while i < L:
        if mask[i]:
            j = i
            while j < L and mask[j]:
                j += 1
            if j - i >= min_len:
                spans.append((i+1, j))  # 1-based inclusive
            i = j
        else:
            i += 1
    return spans

def auto_call_domains(cons_s, kd_s, min_len=12, start_pct=75, end_pct=50, step=5):
    for p in range(start_pct, end_pct-1, -step):
        c_thr = np.percentile(cons_s, p)
        k_thr = np.percentile(kd_s,   p)
        mask = (cons_s >= c_thr) & (kd_s >= k_thr)
        spans = call_domains_mask(mask, min_len=min_len)
        if spans:
            labeled = [(s, e, f"TM{idx+1}") for idx, (s, e) in enumerate(spans)]
            return labeled, dict(percentile=p, cons_thr=float(c_thr), kd_thr=float(k_thr))
    return [], dict(percentile=None, cons_thr=None, kd_thr=None)

domains, meta = auto_call_domains(cons_s, kd_s, min_len=12, start_pct=75, end_pct=50, step=5)
print(f"Auto thresholds -> percentile={meta['percentile']}, cons_thr={meta['cons_thr']}, kd_thr={meta['kd_thr']}")
print("Auto-called domains (start, end, label):", domains)

# ---- candidates (peak picking) ----
score = cons_s + 0.2*kd_s

def top_nonoverlapping_spans(score, k=19, n=4, min_len=14):
    L = len(score)
    taken = np.zeros(L, dtype=bool)
    picks = []
    for _ in range(n):
        sc = np.where(taken, -1e9, score)
        idx = int(sc.argmax())
        if sc[idx] < -1e8:
            break
        peak = score[idx]
        thr  = peak * 0.60
        left = idx
        while left-1 >= 0 and score[left-1] >= thr: left -= 1
        right = idx
        while right+1 < L and score[right+1] >= thr: right += 1
        if (right-left+1) < min_len:
            pad = (min_len - (right-left+1))//2
            left  = max(0, left - pad)
            right = min(L-1, right + pad)
        pad2 = k//2
        lo = max(0, left - pad2); hi = min(L, right + pad2 + 1)
        taken[lo:hi] = True
        picks.append((left+1, right+1))
    picks.sort(key=lambda x: x[0])
    return picks

candidates = top_nonoverlapping_spans(score, k=21, n=4, min_len=14)
print("Candidate TM spans (start, end):", candidates)

# ---- tables ----
def table_for_domains(domains, cons, kd):
    rows = [{
        "Domain": lbl,
        "Start": s,
        "End": e,
        "Length": e - s + 1,
        "Mean_Conservation": float(cons[s-1:e].mean()),
        "Mean_KD":           float(kd[s-1:e].mean())
    } for s, e, lbl in domains]
    return pd.DataFrame(rows, columns=["Domain","Start","End","Length","Mean_Conservation","Mean_KD"])

def table_for_candidates(cands, cons, kd):
    rows = [{
        "Candidate": f"TM{i+1}",
        "Start": s,
        "End": e,
        "Length": e - s + 1,
        "Mean_Conservation": float(cons[s-1:e].mean()),
        "Mean_KD":           float(kd[s-1:e].mean())
    } for i, (s, e) in enumerate(cands)]
    return pd.DataFrame(rows, columns=["Candidate","Start","End","Length","Mean_Conservation","Mean_KD"])

if domains:
    df_domains = table_for_domains(domains, cons, kd)
    print("\nAuto-called TM-like domains:")
    print(df_domains.to_string(index=False))
else:
    print("\nNo auto-called domains with the current auto-thresholds. You can relax them or rely on candidates.")

if candidates:
    df_cands = table_for_candidates(candidates, cons, kd)
    print("\nTop candidate TM helices (peak picking):")
    print(df_cands.to_string(index=False))
else:
    print("\nNo candidate peaks found (unexpected).")
```

```python
# Gets the consensus alignment
def get_consensus(aln):
    cons = []
    for i in range(aln.get_alignment_length()):
        col = [aa for aa in aln[:, i] if aa != "-"]
        if col:
            cons.append(Counter(col).most_common(1)[0][0])
        else:
            cons.append("-")
    return "".join(cons)

# reads in the muscle fasta file to aln
aln = AlignIO.read("PF02932_muscle.fasta", "fasta")

# calulates the consenus of aln
consensus = get_consensus(aln)

# Builds the matrix for the heatmap
matrix = []
for rec in aln:
    row = []
    for a, b in zip(str(rec.seq), consensus):
        if a == b and a != "-":
            row.append(1)
        else:
            row.append(0)
    matrix.append(row)

matrix = np.array(matrix)
print(matrix.shape)  # should be (50, 674)

# plot design
plt.figure(figsize=(10, 6))
plt.imshow(matrix, aspect="auto", cmap="viridis", interpolation="nearest")
plt.colorbar(label="Match to consensus (1=yes, 0=no)")
plt.xlabel("Alignment position")
plt.ylabel("Sequence index")
plt.title("MSA Conservation Consensus Heatmap")
plt.show()
```

  <p style="color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px;">
    Both approaches rely on scoring schemes for matches, mismatches, and gaps to maximize overall alignment score. In both cases, the goal is to detect regions of similarity that may reflect evolutionary relatedness or conserved function. Both types of alignment reveal conserved motifs and variable regions. They are different scales of the same idea: pairwise = two sequences, multiple = many sequences. Both can be used to infer homology and provide input for downstream analysis (e.g. phylogenetics, structure prediction). <br>
      A major difference is that pairwise aligns two sequences directly, while Multiple Sequence Alignment (MSA) aligns many simultaneously.
Pairwise is computationally simpler and exact algorithms are tractable. MSA requires heuristics and approximations because the problem is NP-hard for >3 sequences. Pairwise highlights similarity between a specific pair, while MSA identifies shared conserved residues across an entire protein family, which is crucial for detecting motifs and functional domains.<br><br>
  </p>

  <h4 style="color: #7f8c8d; font-family: 'Georgia', serif; font-weight: normal; font-size: 20px; margin-top: 0; text-align: left;">
    2. What are the core assumptions defining the approach used in each of the BioPython-based tools (e.g. ClustalW, MUSCLE, EMBOSS)?<br>
  </h4>

  <p style="color: #34495e; font-family: 'Arial', sans-serif; font-size: 15px; margin-top: 18px;">
ClustalW assumes that evolutionary relationships can be approximated by a guide tree built from pairwise distances. Clustal aligns sequences progressively according to that tree, so early alignment decisions influence later steps.<br>

MUSCLE assumes that iterative refinement can improve alignment quality. Muscle starts with a quick draft alignment using k-mer distances, then refines repeatedly using updated distance trees and profile alignments. <br>

EMBOSS (Needle, Water) assumes optimal alignment between two sequences can be found with dynamic programming. EMOSS assumes the scoring matrix (e.g. BLOSUM62) accurately reflects evolutionary substitution likelihoods. <br><br>
  </p>
